```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

import csv

url = 'https://www.peakbagger.com/list.aspx?lid=5651'
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')

table = soup.find('table', attrs = {'class': 'gray'})
results = table.find_all('tr')[2:]
#print('Number of results',len(results))

rows = []
rows.append(['no.','name','region','elev','lat','long','url'])
#print(rows)

for result in results:  #per row 
    data = result.find_all('td') 
    print(data)
    if len(data) == 0:
        continue
    no = data[0].text.strip()
    name = data[1].getText()
    region = data[2].getText()
    elev = data[3].getText()

    url2 = data[1].find('a').get('href')
    real_url = 'https://www.peakbagger.com/' + url2
    page = urlopen(real_url).read().decode()
    soup = BeautifulSoup(page,'html.parser')
    n = page.find("(Dec Deg)")
    location = page[n-25:n-1]
    
    lat = location[location.find('>')+1:location.find(',')]
    long = location[location.find(',')+2:]

    rows.append([ no, name, region, elev, lat, long, url2]) 
    print(rows)


with open('../assignment3/3035676171_mountains.csv', 'w') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(rows)
```
